# **5. Diseñe el experimento**

**5.1 Primera politica de reemplazo**

Aquí vamos a generar una muestra de tamaño 2000, en otras palabras esto representa 2000 averias con la primera politica de reemplazamiento, para así obtener el promedio del costo por hora útil de una maquina que sigue esta politica. Después, vamos a obtener la gráfica de estabilidad de esta politica de reeemplazamiento.

```python
def simulaciones (n):
  costos=0  #costos en el tiempo 0
  costo_hora=[] #Arreglo que guarda el costo promedio con la primera politica de reemplazamiento
  for i in range(n):  #Cada iteración es una averia
    costos=costos+descompostura() #costos por hora de vida acumulados por iteración
    costo_hora.append(costos/(i+1)) #Promedio de costos por hora de una maquina en n averias
  return costo_hora
```

```python
x=simulaciones(2000)
```

```python
plt.plot(x,'y')
plt.grid(True)
plt.title('Costos por hora con la primera politica de reemplazo')
plt.ylabel('Costos por hora en euros')
plt.xlabel('Número de descomposturas simuladas en la muestra')
plt.show()
```

```python
print('El costo medio por hora con la primera politica de reemplazamiento es:',np.mean(x),'euros y su desviación estandar:', np.std(x))
```

**5.2 Segunda politica de reemplazo**

Aquí vamos a generar una muestra de tamaño 2000, en otras palabras esto representa 2000 averias con la segunda politica de reemplazamiento, para así obtener el promedio del costo por hora útil de una maquina que sigue esta politica. Después, vamos a obtener la gráfica de estabilidad para esta politica de reemplazamiento.

```python
def simulaciones2 (n):
  costos2=0 #Costos acumulados en el tiempo 0
  costo_hora2=[]
  for i in range(n):  #Cada iteración es una averia
    costos2=costos2+descompostura2()  #costos por hora de vida acumulados de una maquina con la segunda politica de reemplazo
    costo_hora2.append(costos2/(i+1)) #Promedio de costos por hora de una maquina en n averias
  return costo_hora2
```

```python
y=simulaciones2(2000)
```

```python
plt.plot(y,'r')
plt.grid(True)
plt.title('Costos por hora con la segunda politica de reemplazo')
plt.xlabel('Número de averias simuladas')
plt.ylabel('Costos por hora en euros')
plt.show()
```

```python
print('El costo medio por hora con la primera politica de reemplazamiento es:',np.mean(y),'euros y su desviación estandar:', np.std(y))
```

# **6.Calcular N y repetir ambas politicas N veces para comparar cuál es la mejor**
Calculamos N para saber el número de muestras que necesitamos.\

$$ N > \Bigg( \frac{2 \sigma z_{\alpha/2}}{\epsilon} \Bigg) ^2 $$

**6.1 Para la primer politica de reemplazo**

Empezamos por calcular N como el número de muestras.

```python
n= (2*np.std(x)*1.96/0.01)**2 #Calculo de N
print(n)
N=mat.ceil(n)#Redondeamos N al entero de arriba
```

Repetimos el modelo N veces con 2000 averias en cada repeticón.

```python
def modelo(n):
    medias_muestrales=[]
    for i in range(n):  #Repetimos el modelo N veces
        s = simulaciones(2000) #Mandamos a llamar a nuestra función que calcula el costo por hora útil de una maquina con 2000 averias como una muestra
        medias_muestrales.append(np.mean(s))  #Calculamos la media del costo por hora, por cada vez que se ejecutan 2000 averias
    return medias_muestrales  #Lista de tamaño N con las medias muestrales, donde cada muestra es el costo por hora al simular 2000 averias
```

```python
medias= modelo(N) #Obtenemos las N medias muestrales
```

```python
plt.hist(medias, color='y')
plt.grid(True)
plt.xlabel('costos por hora útil')
plt.ylabel('frecuencias')
plt.title('Gráfica de medias muestrales')
plt.show()
```

# **El costo esperado por cada hora de funcionamiento de una maquina con la primera politica de reemplazo es:**

```python
print(np.mean(medias),'euros y con desviación estandar de ', np.std(medias),'euros')
```

**6.2 Para la segunda politica de reemplazo**

```python
n2= (2*np.std(y)*1.96/0.01)**2 #Calculo de N
print(n2)
N2=mat.ceil(n2)#Redondeamos N al entero de arriba
```

Repetimos el modelo N veces con 2000 averias en cada repeticón.

```python
def modelo2(n):
    medias_muestrales2=[]
    for i in range(n):  #Repetimos el modelo N veces
        s2 = simulaciones2(2000) #Mandamos a llamar a nuestra función que calcula el costo por hora útil de una maquina con 1000 averias como una muestra
        medias_muestrales2.append(np.mean(s2))  #Calculamos la media del costo por hora, por cada vez que se ejecutan 1000 averias
    return medias_muestrales2  #Lista de tamaño N con las medias muestrales, donde cada muestra es el costo por hora al simular 1000 averias
```

```python
medias2=modelo2(N2)
```

```python
plt.hist(medias2, color='r',bins=10)
plt.grid(True)
plt.xlabel('costos por hora útil')
plt.ylabel('frecuencias')
plt.title('Gráfica de medias muestrales')
plt.show()
```

## **7.Hallar intervalos de confianza**
\
$$\bar{x}-z_{\alpha/2} \cdot \frac{s}{\sqrt{n}} < \mu < \bar{x}+z_{\alpha/2}\cdot \frac{s}{\sqrt{n}}$$

**7.1 Intervalos de confianza para el modelo de la primer politica de reemplazo**

```python
np.mean(medias) - 1.96*np.std(medias)/np.sqrt(N), np.mean(medias) + 1.96*np.std(medias)/np.sqrt(N)
```

```python
np.mean(medias2) - 1.96*np.std(medias2)/np.sqrt(N2), np.mean(medias2) + 1.96*np.std(medias2)/np.sqrt(N2)
```
