# **5. Disee el experimento**

**5.1 Primera politica de reemplazo**

Aqu vamos a generar una muestra de tamao 2000, en otras palabras esto representa 2000 averias con la primera politica de reemplazamiento, para as obtener el promedio del costo por hora til de una maquina que sigue esta politica. Despus, vamos a obtener la grfica de estabilidad de esta politica de reeemplazamiento.

```python
def simulaciones (n):
  costos=0  #costos en el tiempo 0
  costo_hora=[] #Arreglo que guarda el costo promedio con la primera politica de reemplazamiento
  for i in range(n):  #Cada iteracin es una averia
    costos=costos+descompostura() #costos por hora de vida acumulados por iteracin
    costo_hora.append(costos/(i+1)) #Promedio de costos por hora de una maquina en n averias
  return costo_hora
```

```python
x=simulaciones(2000)
```

```python
plt.plot(x,'y')
plt.grid(True)
plt.title('Costos por hora con la primera politica de reemplazo')
plt.ylabel('Costos por hora en euros')
plt.xlabel('Nmero de descomposturas simuladas en la muestra')
plt.show()
```

```python
print('El costo medio por hora con la primera politica de reemplazamiento es:',np.mean(x),'euros y su desviacin estandar:', np.std(x))
```

**5.2 Segunda politica de reemplazo**

Aqu vamos a generar una muestra de tamao 2000, en otras palabras esto representa 2000 averias con la segunda politica de reemplazamiento, para as obtener el promedio del costo por hora til de una maquina que sigue esta politica. Despus, vamos a obtener la grfica de estabilidad para esta politica de reemplazamiento.

```python
def simulaciones2 (n):
  costos2=0 #Costos acumulados en el tiempo 0
  costo_hora2=[]
  for i in range(n):  #Cada iteracin es una averia
    costos2=costos2+descompostura2()  #costos por hora de vida acumulados de una maquina con la segunda politica de reemplazo
    costo_hora2.append(costos2/(i+1)) #Promedio de costos por hora de una maquina en n averias
  return costo_hora2
```

```python
y=simulaciones2(2000)
```

```python
plt.plot(y,'r')
plt.grid(True)
plt.title('Costos por hora con la segunda politica de reemplazo')
plt.xlabel('Nmero de averias simuladas')
plt.ylabel('Costos por hora en euros')
plt.show()
```

```python
print('El costo medio por hora con la primera politica de reemplazamiento es:',np.mean(y),'euros y su desviacin estandar:', np.std(y))
```

# **6.Calcular N y repetir ambas politicas N veces para comparar cul es la mejor**
Calculamos N para saber el nmero de muestras que necesitamos.\

$$ N > \Bigg( \frac{2 \sigma z_{\alpha/2}}{\epsilon} \Bigg) ^2 $$

**6.1 Para la primer politica de reemplazo**

Empezamos por calcular N como el nmero de muestras.

```python
n= (2*np.std(x)*1.96/0.01)**2 #Calculo de N
print(n)
N=mat.ceil(n)#Redondeamos N al entero de arriba
```

Repetimos el modelo N veces con 2000 averias en cada repeticn.

```python
def modelo(n):
    medias_muestrales=[]
    for i in range(n):  #Repetimos el modelo N veces
        s = simulaciones(2000) #Mandamos a llamar a nuestra funcin que calcula el costo por hora til de una maquina con 2000 averias como una muestra
        medias_muestrales.append(np.mean(s))  #Calculamos la media del costo por hora, por cada vez que se ejecutan 2000 averias
    return medias_muestrales  #Lista de tamao N con las medias muestrales, donde cada muestra es el costo por hora al simular 2000 averias
```

```python
medias= modelo(N) #Obtenemos las N medias muestrales
```

```python
plt.hist(medias, color='y')
plt.grid(True)
plt.xlabel('costos por hora til')
plt.ylabel('frecuencias')
plt.title('Grfica de medias muestrales')
plt.show()
```

# **El costo esperado por cada hora de funcionamiento de una maquina con la primera politica de reemplazo es:**

```python
print(np.mean(medias),'euros y con desviacin estandar de ', np.std(medias),'euros')
```

## **7.Hallar intervalos de confianza**
\
$$\bar{x}-z_{\alpha/2} \cdot \frac{s}{\sqrt{n}} < \mu < \bar{x}+z_{\alpha/2}\cdot \frac{s}{\sqrt{n}}$$

**7.1 Intervalos de confianza para el modelo de la primer politica de reemplazo**

```python
np.mean(medias) - 1.96*np.std(medias)/np.sqrt(N), np.mean(medias) + 1.96*np.std(medias)/np.sqrt(N)
```
